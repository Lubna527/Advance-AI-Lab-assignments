```python
from collections import deque

def BFS(a, b, target):
    # Dictionary to mark visited states
    visited = {}
    # Flag to check if the solution is found
    isSolvable = False
    # List to store the path to the solution
    path = []

    # Queue for BFS starting with both jugs empty
    q = deque()
    q.append((0, 0))

    while q:
        current_state = q.popleft()

        # If this state is already visited, skip it
        if current_state in visited:
            continue

        # Extract current jug levels
        jug1, jug2 = current_state

        # Skip invalid states (outside jug capacities)
        if jug1 > a or jug2 > b or jug1 < 0 or jug2 < 0:
            continue

        # Record this state as part of the solution path
        path.append([jug1, jug2])

        # Mark this state as visited
        visited[current_state] = True

        # Print the current state
        print(f"Current state: Jug 1 = {jug1} liters, Jug 2 = {jug2} liters")

        # Check if we've reached the target in either jug
        if jug1 == target or jug2 == target:
            isSolvable = True
            print("\nSolution found!")
            if jug1 == target and jug2 != 0:
                path.append([jug1, 0])
            if jug2 == target and jug1 != 0:
                path.append([0, jug2])
            
            break

        # Generate possible next states:
        q.append((jug1, b))  # Fill Jug 2
        print(f"Fill Jug 2: Jug 1 = {jug1} liters, Jug 2 = {b} liters")
        
        q.append((a, jug2))  # Fill Jug 1
        print(f"Fill Jug 1: Jug 1 = {a} liters, Jug 2 = {jug2} liters")
        
        # Pour from Jug 2 to Jug 1
        for pour_amount in range(max(a, b) + 1):
            # Pour from Jug 2 to Jug 1
            new_jug1 = jug1 + pour_amount
            new_jug2 = jug2 - pour_amount

            if new_jug1 == a or (new_jug2 == 0 and new_jug2 >= 0):
                q.append((new_jug1, new_jug2))
                print(f"Pour from Jug 2 to Jug 1: Jug 1 = {new_jug1} liters, Jug 2 = {new_jug2} liters")

            # Pour from Jug 1 to Jug 2
            new_jug1 = jug1 - pour_amount
            new_jug2 = jug2 + pour_amount

            if (new_jug1 == 0 and new_jug1 >= 0) or new_jug2 == b:
                q.append((new_jug1, new_jug2))
                print(f"Pour from Jug 1 to Jug 2: Jug 1 = {new_jug1} liters, Jug 2 = {new_jug2} liters")

        # Empty Jug 2
        q.append((jug1, 0))
        print(f"Empty Jug 2: Jug 1 = {jug1} liters, Jug 2 = 0 liters")

        # Empty Jug 1
        q.append((0, jug2))
        print(f"Empty Jug 1: Jug 1 = 0 liters, Jug 2 = {jug2} liters")

    # Print the final path if the solution is found
    if isSolvable:
        print("\nPath to solution:")
        for state in path:
            print(f"Jug 1 = {state[0]} liters, Jug 2 = {state[1]} liters")
    else:
        print("No solution found")

# Driver code
if __name__ == '__main__':
    Jug1, Jug2, target = 4, 3, 2
    # Get user input for jug capacities and target amount
    #Jug1 = int(input("Enter the capacity of Jug 1: "))
    #Jug2 = int(input("Enter the capacity of Jug 2: "))
    #target = int(input("Enter the target amount of water: "))
    print("Starting BFS to find solution...\n")
    BFS(Jug1, Jug2, target)
```

```python
from collections import deque

def is_valid_state(state, jug1_capacity, jug2_capacity, visited):
    jug1, jug2 = state
    return 0 <= jug1 <= jug1_capacity and 0 <= jug2 <= jug2_capacity and state not in visited

def water_jug_solver_all_solutions(jug1_capacity, jug2_capacity, goal):
    initial_state = (0, 0)
    visited = set()
    queue = deque([(initial_state, [])])
    solutions = []

    while queue:
        (jug1, jug2), path = queue.popleft()

        if (jug1, jug2) in visited:
            continue

        visited.add((jug1, jug2))

        if jug1 == goal or jug2 == goal or jug1 + jug2 == goal:
            solutions.append(path + [(jug1, jug2)])

        next_states = [
            (jug1_capacity, jug2),  # Fill Jug 1
            (jug1, jug2_capacity),  # Fill Jug 2
            (0, jug2),              # Empty Jug 1
            (jug1, 0),              # Empty Jug 2
            (jug1 - min(jug1, jug2_capacity - jug2), jug2 + min(jug1, jug2_capacity - jug2)),  # Pour Jug 1 -> Jug 2
            (jug1 + min(jug2, jug1_capacity - jug1), jug2 - min(jug2, jug1_capacity - jug1))   # Pour Jug 2 -> Jug 1
        ]

        for state in next_states:
            if is_valid_state(state, jug1_capacity, jug2_capacity, visited):
                queue.append((state, path + [(jug1, jug2)]))

    return solutions if solutions else "No solution found"

# Take inputs from the user
jug1_capacity = int(input("Enter the capacity of Jug 1: "))
jug2_capacity = int(input("Enter the capacity of Jug 2: "))
goal = int(input("Enter the goal amount of water: "))

solutions = water_jug_solver_all_solutions(jug1_capacity, jug2_capacity, goal)

print("All possible solutions to reach the goal:")
if isinstance(solutions, list):
    for index, solution in enumerate(solutions, start=1):
        print(f"Solution {index}:")
        for step in solution:
            print(step)
else:
    print(solutions)
```
