```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Define a fixed set of 5 cities with specific coordinates for reproducibility
def generate_5_cities_dict():
    return {
        "City_A": np.array([0.1, 0.2]),
        "City_B": np.array([0.4, 0.9]),
        "City_C": np.array([0.9, 0.3]),
        "City_D": np.array([0.2, 0.8]),
        "City_E": np.array([0.7, 0.5])
    }

# Calculate the total distance of the route
def total_distance(route, cities):1
    turn sum(np.linalg.norm(cities[route[i]] - cities[route[i-1]]) for i in range(len(route)))

# Visualization function for a route
def plot_route(route, cities, title="Route"):
    plt.figure(figsize=(8, 6))
    coords = np.array([cities[city] for city in route] + [cities[route[0]]])  # Append starting city to close loop
    plt.plot(coords[:, 0], coords[:, 1], '-o', markersize=5)
    plt.title(title)
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.show()

# Simulated Annealing Algorithm
def simulated_annealing(cities, initial_temp=1000, cooling_rate=0.003, max_iterations=1000):
    cities_list = list(cities.keys())
    current_route = cities_list.copy()
    random.shuffle(current_route)
    current_distance = total_distance(current_route, cities)
    
    best_route = current_route
    best_distance = current_distance
    temp = initial_temp
    
    for i in range(max_iterations):
        # Generate a neighboring solution by swapping two cities
        new_route = current_route.copy()
        idx1, idx2 = np.random.randint(0, len(cities_list), 2)
        new_route[idx1], new_route[idx2] = new_route[idx2], new_route[idx1]
        
        # Calculate distance of new route
        new_distance = total_distance(new_route, cities)
        
        # Accept new route if it's better or with some probability if it's worse
        if new_distance < current_distance or np.random.rand() < np.exp((current_distance - new_distance) / temp):
            current_route = new_route
            current_distance = new_distance
            
            # Update best route found
            if current_distance < best_distance:
                best_route = current_route
                best_distance = current_distance
        
        # Cool down
        temp *= (1 - cooling_rate)
    
    return best_route

# Test and visualize Simulated Annealing with 5 cities
cities = generate_5_cities_dict()
best_route_sa = simulated_annealing(cities)
plot_route(best_route_sa, cities, title="Simulated Annealing")

print("Total distance using Simulated Annealing:", total_distance(best_route_sa, cities))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Define a fixed set of 5 cities with specific coordinates for reproducibility
def generate_5_cities_dict():
    return {
        "City_A": np.array([0.1, 0.2]),
        "City_B": np.array([0.4, 0.9]),
        "City_C": np.array([0.9, 0.3]),
        "City_D": np.array([0.2, 0.8]),
        "City_E": np.array([0.7, 0.5])
    }

# Calculate the total distance of the route
def total_distance(route, cities):
    return sum(np.linalg.norm(cities[route[i]] - cities[route[i-1]]) for i in range(len(route)))

# Visualization function for a route
def plot_route(route, cities, title="Route"):
    plt.figure(figsize=(8, 6))
    coords = np.array([cities[city] for city in route] + [cities[route[0]]])  # Append starting city to close loop
    plt.plot(coords[:, 0], coords[:, 1], '-o', markersize=5)
    plt.title(title)
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.show()

# Genetic Algorithm Functions

# Create initial population
def create_population(cities, pop_size):
    cities_list = list(cities.keys())
    return [random.sample(cities_list, len(cities_list)) for _ in range(pop_size)]

# Calculate fitness of an individual route
def fitness(route, cities):
    return 1 / total_distance(route, cities)  # Inverse distance

# Selection of individuals based on fitness (tournament selection)
def select(population, cities):
    tournament = random.sample(population, 3)
    return min(tournament, key=lambda route: total_distance(route, cities))

# Crossover function to create a child from two parent routes
def crossover(parent1, parent2):
    start, end = sorted(random.sample(range(len(parent1)), 2))
    child = [None] * len(parent1)
    child[start:end] = parent1[start:end]
    
    ptr = end
    for gene in parent2:
        if gene not in child:
            if ptr >= len(parent1):
                ptr = 0
            child[ptr] = gene
            ptr += 1
    return child

# Mutation function to introduce variation in the route
def mutate(route, mutation_rate=0.01):
    for i in range(len(route)):
        if random.random() < mutation_rate:
            j = random.randint(0, len(route) - 1)
            route[i], route[j] = route[j], route[i]

# Genetic Algorithm function
def genetic_algorithm(cities, pop_size=10, generations=50, mutation_rate=0.01):
    population = create_population(cities, pop_size)
    best_route = min(population, key=lambda route: total_distance(route, cities))
    
    for _ in range(generations):
        # Select next generation
        next_population = []
        while len(next_population) < pop_size:
            parent1 = select(population, cities)
            parent2 = select(population, cities)
            child = crossover(parent1, parent2)
            mutate(child, mutation_rate)
            next_population.append(child)
        
        # Update the population and the best route
        population = next_population
        current_best_route = min(population, key=lambda route: total_distance(route, cities))
        if total_distance(current_best_route, cities) < total_distance(best_route, cities):
            best_route = current_best_route
            
    return best_route

# Test and visualize Genetic Algorithm with 5 cities
cities = generate_5_cities_dict()
best_route_ga = genetic_algorithm(cities)
plot_route(best_route_ga, cities, title="Genetic Algorithm")

print("Total distance using Genetic Algorithm:", total_distance(best_route_ga, cities))
```
